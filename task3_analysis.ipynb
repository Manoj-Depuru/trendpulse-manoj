```python
import pandas as pd
import numpy as np
# Loading .csv file
df=pd.read_csv("trends_clean.csv")
# printing the first 5 rows
print("The first 5 rows:")
print(df.head())
# printing the shape
print("\n Shape of data set:",df.shape)
# Average of score and num_comments
print("\n Average score:",df["score"].mean())
print("\n Average num_comments:",df["num_comments"].mean())
```

Task-2


```python
scores=df["score"].values
print("\n Mean:",np.mean(scores))
print(" Median:",np.median(scores))
print("Standard Deviation:",np.std(scores))
print("\n Highest score:",np.max(scores))
print("Lowest score:",np.min(scores))
# Story per catgeory
counts=df["category"].value_counts()
Highest_category=counts.idxmax()
highest_count=counts.max()
print(f"Category with most stories:{Highest_category}({highest_count})")
# Story with most comments
max_comments=df.loc[df["num_comments"].idxmax()]
print("\n Story with most comments is:")
print("Title:",max_comments["title"])
print("Comments:",max_comments["num_comments"])
```

Task-3


```python
# Adding new colums
df["engagement"]=df["num_comments"] / (df["score"]+1)
average_score=df["score"].mean()
df["is_popular"]=df["score"]>average_score
```

Task-4

```python
df.to_csv("trends_analysed.csv",index=False)
print("\n File saved to trends_analysed.csv")
```
